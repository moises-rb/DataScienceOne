```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar uma sequncia de valores de x de -1 a 1
x = np.arange(-1, 1.01, 0.0001)

# Implementao da frmula
y1 = np.sqrt(1 - x**2)
y2 = -np.sqrt(1 - x**2)

# Plotar o grfico com as duas partes do crculo
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'r' )

# Adicionar o ttulo do grfico e os rtulos dos eixos x e y
plt.title("Crculo")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")

# Exibir o grfico
plt.show()
```
